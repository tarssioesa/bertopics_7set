Chamando os pacotes que serÃ£o utilizados:

```python
!pip install bertopic
!pip install sentence_transformers
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.feature_extraction.text import CountVectorizer
from bertopic import BERTopic
```

Abrindo nosso arquivo e limpando de linhas vazias:

```python
### transform in pandas:

df = pd.read_csv("/content/bertopics_7set/7set_2021.csv")

### dic para filtrar

df2 = df[~df['keyText'].isnull()]

### Definido docs

docs = df2['keyText']

def remove_duplicates(text_vector):
    unique_texts = set(text_vector)
    unique_text_vector = list(unique_texts)
    return unique_text_vector

docs = remove_duplicates(docs)


```

Rodando bertopics:

```python
### usando embeddings:

from bertopic.representation import KeyBERTInspired
from bertopic.vectorizers import ClassTfidfTransformer
from umap import UMAP

### settings:

embedding_model = SentenceTransformer("intfloat/multilingual-e5-large")
vectorizer_model = CountVectorizer(stop_words="portuguese")
representation_model = KeyBERTInspired()

# Use the representation model in BERTopic on top of the default pipeline
# Train our topic model using our pre-trained sentence-transformers embeddings
topic_model = BERTopic(embedding_model=embedding_model,
                       nr_topics = 15,
                       representation_model= representation_model)

topics, probs = topic_model.fit_transform(docs)

```

Nossos resultados:

```python
# Frequencia dos topicos:

freq = topic_model.get_topic_info()

freq
```

```python
# textos representativos
df = pd.DataFrame({'topic': topics, 'document': docs})


with open('/content/bertopics_7set/resultados.txt', 'w') as f:
    # Loop over the topics and write the output to the file
    for topic in range(13):
        topic_info = topic_model.get_topic(topic)
        representative_docs = df[df.topic == topic][:5]['document'].tolist()

        f.write(f"Topic: {topic}\n")
        f.write("Topic Information: \n")
        f.write(f"{topic_info}\n")
        f.write("Representative Documents: \n")
        for doc in representative_docs:
            f.write(f"{doc}\n")
        f.write("\n")
```

```python
resultado = pd.DataFrame({'topic': topics, 'document': docs})

resultado.to_csv('/content/bertopics_7set/resultados.csv', index=False)
```
